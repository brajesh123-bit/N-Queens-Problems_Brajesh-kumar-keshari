def is_safe(board, row, col, n): ““” Checks if it’s safe to place a
queen at position (row, col) A position is safe if: - No queen exists in
the same row on the left side - No queen exists in the upper left
diagonal - No queen exists in the lower left diagonal ““”

    # Check this row on the left side
    for i in range(col):
        if board[row][i]:
            return False

    # Check upper diagonal on the left side
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j]:
            return False

    # Check lower diagonal on the left side
    for i, j in zip(range(row, n, 1), range(col, -1, -1)):
        if board[i][j]:
            return False

    return True
